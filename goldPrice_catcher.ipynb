```python
# 匯入套件
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. 設定下載代號 (GC=F 是黃金期貨, GLD 是黃金ETF)
gold_ticker = "GC=F"

# 2. 下載過去兩年的歷史資料
gold_data = yf.download(gold_ticker, start="2024-01-01", end="2026-01-28")

# 3. 計算 20 日與 50 日移動平均線 (幫助觀察趨勢)
gold_data['MA20'] = gold_data['Close'].rolling(window=20).mean()
gold_data['MA50'] = gold_data['Close'].rolling(window=50).mean()
```

```python
# 4. 繪製圖表
plt.figure(figsize=(14, 7))
plt.plot(gold_data['Close'], label='Gold Price (Daily Close)', color='gold', alpha=0.6)
plt.plot(gold_data['MA20'], label='20-Day MA', color='blue', linestyle='--')
plt.plot(gold_data['MA50'], label='50-Day MA', color='red', linestyle='--')

plt.title('Gold Price Trend Analysis (2024 - 2026)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 顯示最後幾筆資料，確認數據沒問題
print(gold_data.tail())
```
